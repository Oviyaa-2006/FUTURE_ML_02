```python
import joblib

# Load vectorizer
vectorizer = joblib.load("count_vectorizer.pkl")

# Load models
topic_model = joblib.load("random_forest_final_modelTopic Classification.pkl")
priority_model = joblib.load("random_forest_final_modelPriority Classificatoin.pkl")
```

```python
new_ticket = "Please reset my password as I cannot log in."

# Convert text to Bag-of-Words features
X_new = vectorizer.transform([new_ticket])

# Predict
topic = topic_model.predict(X_new)[0]
priority = priority_model.predict(X_new)[0]

print("Topic:", topic)
print("Priority:", priority)
```

```python
tickets = [
    "I am unable to log in to my account. Please reset my password.",
    "My laptop is not powering on after pressing the power button.",
    "Microsoft Outlook keeps crashing whenever I open it.",
    "Please update my leave balance in the HR portal.",
    "The office printer is not working.",
    "Urgent need of administrator rights to install software.",
    "Request access to shared folder.",
    "Add me to the XYZ project team.",
    "Please assist me with this service request.",
    "Request approval for purchasing a new monitor."
]

X_new = vectorizer.transform(tickets)

topic_predictions = topic_model.predict(X_new)
priority_predictions = priority_model.predict(X_new)

for ticket, topic, priority in zip(tickets, topic_predictions, priority_predictions):
    print(f"Ticket: {ticket}")
    print(f"Topic: {topic}")
    print(f"Priority: {priority}")
    print("-" * 50)
```

```python
while True:
    ticket = input("\nEnter a support ticket (or type 'exit' to quit): ")

    if ticket.lower() == "exit":
        print("Exiting...")
        break

    # ticket = clean_text(ticket)  # Uncomment if needed

    X_new = vectorizer.transform([ticket])

    topic = topic_model.predict(X_new)[0]
    priority = priority_model.predict(X_new)[0]

    print("\nPrediction")
    print(f"Topic    : {topic}")
    print(f"Priority : {priority}")
```
